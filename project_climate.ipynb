```python
def FindMissingColsPercentage(df):

    total = 0
    for col in df.columns:
        missing_vals = df[col].isnull().sum()
        # mean = sum / total
        pct = df[col].isnull().mean() * 100 
        if missing_vals != 0:
          print('{} => {} [{}%]'.format(col, df[col].isnull().sum(), round(pct, 2)))
        total += missing_vals
    
    if total == 0:
        print("no missing values")
```

1. How has the average temperature in Windsor region changed over the past century? Is there a clear trend?

```python
# Create a new dataframe with only the Windsor region data
windsor_df = df[df['Station_Name'].str.contains('WINDSOR')]

# Group by year and calculate the mean temperature
mean_temp_by_year = windsor_df.groupby('Year')['Mean_Temp_'].mean()

# Create a line plot of mean temperature by year
plt.plot(mean_temp_by_year.index, mean_temp_by_year, label = 'Temperature (°C)' )
plt.xlabel('Year')
plt.ylabel('Mean Temperature (°C)')
plt.title('Mean Temperature in Windsor Region (2013-2022)', color = 'navy')
plt.tick_params(axis='both', which='major', labelsize=12)
plt.tight_layout()
plt.legend()
plt.grid()
plt.show()
```

2. How have extreme precipitation events (measured by total precipitation) in Windsor region changed over time?

```python
# Filter the dataset to only include Windsor region data
windsor_df = df[df['Station_Name'].str.contains('WINDSOR')]

# Calculate the total precipitation by year
total_precip_by_year = windsor_df.groupby('Year')['Total_Precip_'].sum()

# Set the color palette
colors = sns.color_palette("Blues_r", len(total_precip_by_year))

# Create a bar plot of total precipitation by year
plt.bar(total_precip_by_year.index, total_precip_by_year, hatch = '/', label = 'Total Precip (mm)')
plt.xlabel('Year')
plt.ylabel('Total Precipitation (mm)')
plt.title('Total Precipitation in Windsor Region (2013-2022)')

# Customize plot
plt.xticks(total_precip_by_year.index, rotation=45)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.yticks(ticks=[0, 400, 800, 1200, 1600, 2000, 2400, 2800, 3200])
plt.tight_layout()
plt.legend()
plt.grid()
plt.show()
```

3. How has the mean temperature in Windsor region changed over the months with respect to the years, and what are the seasonal patterns in temperature? 

```python
# Filter the dataset to only include Windsor region data
windsor_df = df[df['Station_Name'].str.contains('WINDSOR')]

# Calculate the mean temperature by year and month
mean_temp_by_month = windsor_df.groupby(['Year', 'Month'])['Mean_Temp_'].mean().reset_index()

# Pivot the data to have years as rows and months as columns
pivot_mean_temp = mean_temp_by_month.pivot(index='Year', columns='Month', values='Mean_Temp_')

# Create a heatmap of mean temperature by year and month
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_mean_temp, cmap='coolwarm', annot=True, fmt=".1f", cbar_kws={'label': 'Mean Temperature (°C)'})
plt.title('Mean Temperature by Year and Month in Windsor Region (2013-2022)')
plt.show()
```

5. How has the maximum wind gust speed in Windsor region varied with mean temperature from the last 5 years?

```python
# Replace '<31' values in 'Spd_of_Max_Gust' column with NaN
windsor_df['Spd_of_Max_Gust'] = windsor_df['Spd_of_Max_Gust'].replace('<31', np.nan)

# Convert 'Spd_of_Max_Gust' column to float64
windsor_df['Spd_of_Max_Gust'] = windsor_df['Spd_of_Max_Gust'].astype('float64')

# Filter data for the years 2017-2022
yearly_data = windsor_df[windsor_df['Year'].between(2017, 2022)].groupby('Year').agg({'Mean_Temp_': 'min', 'Spd_of_Max_Gust': 'max'}).reset_index()

#Create a bubble plot
fig, ax = plt.subplots(figsize=(8,6))
sc = ax.scatter(yearly_data['Mean_Temp_'], yearly_data['Spd_of_Max_Gust'], s=yearly_data['Year']*5, alpha=0.5, c=yearly_data['Year'], cmap='cool')
plt.xlim(yearly_data['Mean_Temp_'].min() - 1, yearly_data['Mean_Temp_'].max() + 1)
plt.ylim(yearly_data['Spd_of_Max_Gust'].min() - 10, yearly_data['Spd_of_Max_Gust'].max() + 10)
plt.xlabel('Mean Temperature (°C)')
plt.ylabel('Maximum Wind Gust Speed (km/h)')
plt.title('Maximum Wind Gust Speed vs. Mean Temperature for the last 5 years in Windsor Region')

#Create legend
legend1 = ax.legend(*sc.legend_elements(num=6), title="Year", loc="upper left", bbox_to_anchor=(1.05, 1), markerscale=0.5)

#Add annotation to each data point
for i, year in enumerate(yearly_data['Year']):
    plt.annotate(year, (yearly_data['Mean_Temp_'][i], yearly_data['Spd_of_Max_Gust'][i]), ha='center')

plt.grid()
plt.show()

```
